```python
#esta función retorna la lista de compuestos que interactuan con una celula
def get_listaCompuestos(cell):
    lista_compuestos = perfil_afinidad_activity['id_compuestos_actividad'][perfil_afinidad_activity['cell_id']==cell].tolist()
    try:
        lista_compuestos = eval(lista_compuestos[0]) #esto es un list to string
    except:
        pass
    return lista_compuestos
```

```python
#esta función retorna la lista de compuestos que interactuan con una celula, además de la actividad
def get_listaCompuestos_conActividad(cell):
    try:
        lista_compuestos_actividad = [eval(perfil_afinidad_activity['id_compuestos_actividad'][perfil_afinidad_activity['cell_id']==cell].tolist()[0]),eval(perfil_afinidad_activity['activity'][perfil_afinidad_activity['cell_id']==cell].tolist()[0])]
    except:
        lista_compuestos_actividad = []
    return lista_compuestos_actividad
```

```python
#esta funcion retornar el top 3 de un compuesto, con los ids pertenecientes a chembl
def get_top3(compound):
    top3_id29 = []
    top3_compuesto = top3['top3'][top3['id_29'] == compound].tolist()
    if top3_compuesto:
        top3_compuesto = eval(top3_compuesto[0])
        for i in top3_compuesto:
            test = top3['id_29'][top3['comp_id'] == i]
            if not test.empty:
                top3_id29.append(test)
        if top3_id29:
            for i in range(len(top3_id29)):
                top3_id29[i] = top3_id29[i].tolist()
    return top3_id29
```

```python
def get_top3_29(compuesto):
    top3_compuesto = []
    try:
        top3_compuesto = eval(top3['top3_29'][top3['id_29'] == compuesto].tolist()[0])
    except:
        pass
    return top3_compuesto
```
